```python
import matplotlib.pyplot as plt
from keras.src.layers import LSTM
from plot_keras_history import plot_history

from study.a_rec import ActivityRecognitionModel
from util.vis import get_model_vis
```

```python
model = ActivityRecognitionModel(segment_size=50, batch_size=200, n_shifts=4)

model.add_input_layer()
model.add(LSTM(700, activation="tanh"))
model.add_output_layer()

model.compile(
    optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"]
)

get_model_vis(model)
```

```python
history = model.train_model()
plt.show(plot_history(history))
```

```python
model.evaluate_model()
```
