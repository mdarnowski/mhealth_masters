```python
import matplotlib.pyplot as plt
from keras.src.layers import LSTM
from keras.src.utils import plot_model
from plot_keras_history import plot_history

from models.a_rec import ActivityRecognitionModel
```

```python
model = ActivityRecognitionModel(segment_size=50, batch_size=50, n_shifts=3)

model.add_input_layer()
model.add(LSTM(200, activation="tanh"))
model.add_output_layer()

model.compile(
    optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"]
)

plot_model(model, show_shapes=True, show_layer_activations=True, dpi=100)
```

```python
history = model.train_model()
plt.show(plot_history(history))
```

```python
model.evaluate_model()
```
