```python
from study.a_rec import ActivityRecognitionModel
```

```python
from keras.src.layers import GRU

from util import vis
```

```python
model = ActivityRecognitionModel(
    segment_size=50,
    batch_size=100,
    n_shifts=1,
    epochs=15,
    l_rate=0.001,
    optimizer_name="adam",
)

model.add_input_layer()
model.add(GRU(128))
model.add_output_layer()

model.compile_model()

vis.get_model_vis(model)
```

```python
history = model.train_model()
```

```python
vis.plot_val_history(history, "Model Training History")
```

```python
vis.plot_evaluation_results(model, history, "Model Evaluation Results")
```
